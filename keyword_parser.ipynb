```python
def Parse (path):
    arr=[]
    cwd = os.getcwd()
    file_path = os.path.join(cwd, path)
    df = pd.DataFrame()
    f = open(file_path,'rt', encoding='UTF8')
    
    df = pd.DataFrame(columns = ['year', 'month', 'day', 'name', 'time', 'content'])
    year=0
    month=0
    day=0
    hour=0
    date=""
    for line in f.readlines():
        if line.startswith('------'):
            date = line.lstrip(' ')
            date = line.rstrip('\n')
            date = date.strip('---------------')
            date = date.strip(' ')
            d = date.split(' ')
            d[2] = d[2].rstrip("일")
            if len(d[2])==1:
                d[2]='0'+d[2]
            date = d[0].rstrip('년')+" "+d[1].rstrip('월')+" "+d[2]
            year = int(d[0].rstrip('년'))
            month = d[1].rstrip('월')
            day = d[2]
#             time = d[4].split(':')
#             if d[3] == "오후":
#                 time[0]=int(time[0])+12
#             d[4]=time[0]+":"+time[1]
#             print(date)
        elif line.startswith('['): #사람의 이름을 넣어야함.
            l = line.split(']',2)
            if(len(l)<3):
                continue
            
            time = l[1].lstrip(' [')
            time = time.split(" ")
            t = time[1].split(":")
            tttt= ""
            if(time[0]=="오후"):
                t[0] = int(t[0])+12

            
            tttt = str(t[0])+":"+t[1]
            if(int(t[0])<10):
                tttt = '0'+tttt
            
            #df = df.append({'year':year,'month':month,'day':day,'name',l[0],'time':tttt,'content':l[2].strip('\n')})
            df.loc[len(df)] = [year,month,day,l[0].lstrip('['),tttt,l[2].strip('\n')]

    return  df
       
```

```python
def getDialogAsText(start, end, dataframe):
    df = dataframe.iloc[start:end]
    text = ""
    for idx in df.index:
        d = df.loc[idx]
        text+= d['content'] + '\n'
        
    return text
```
